```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
# %matplotlib inline
# from google.colab.patches import cv2_imshow
# import pywt
# import pywt.data
from math import sqrt
from PIL import Image, ImageChops, ImageOps
import pickle
```

```python
#1-st and 2-nd task
img = cv2.cvtColor(cv2.imread('img_1.jpg'), cv2.COLOR_BGR2GRAY) # ur img here
ft = np.fft.fft2(img)
comp = ft[521][1151]
print('Введите действительную часть полученного комплексного числа:', round(comp.real, 2), '\n',
      'Введите мнимую часть полученного комплексного числа:',  round(comp.imag, 2))
```

```python
#3nd 
ftshift = np.fft.fftshift(ft)
magnitude_spectrum = np.abs(ftshift)
print('Определите минимальное значение амплитуды:',round(magnitude_spectrum.min(), 2), '\n',
      'Определите максимальное значение амплитуды:', round(magnitude_spectrum.max(), 2))
```

```python
#4th
magnitude_spectrum = np.log(1+np.abs(ftshift))
print('Определите различие в полученных значениях (разность между наибольшим и наименьшим):', round(magnitude_spectrum.max() - magnitude_spectrum.min(), 2))
```

```python
#5th
h, w = img.shape
cy, cx = h // 2, w // 2
ftshift[cy-40:cy+40, cx-40:cx+40] = 0
ft_ishift = np.fft.ifftshift(ftshift)
img_back = np.fft.ifft2(ft_ishift)
img_back = np.abs(img_back)
print('Определите амплитуду обратного преобразования',round(img_back[661][449], 2)) #ваши пиксели сюда 
```
