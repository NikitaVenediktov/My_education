```python
import cv2
import matplotlib.pyplot as plt
import os

img = cv2.imread('1.jpg')
print(img.shape) #1-высота, 2-ширина

print('Отношение:', img.shape[1]/img.shape[0]) 
print('ширина:', img.shape[1]) 
print('Высота:', img.shape[0]) 
```

```python
h, w, c = img.shape
new_width = 1967

k = float(w) / h
dim = (new_width, int(new_width / k))

resized_RGB_img = cv2.resize(img, dim, interpolation = cv2.INTER_AREA)

print('Отношение:', resized_RGB_img.shape[1]/resized_RGB_img.shape[0]) 
print('ширина:', resized_RGB_img.shape[1]) 
print('Высота:', resized_RGB_img.shape[0]) 
plt.imshow(resized_RGB_img)
print(resized_RGB_img.shape) # новая размерность 
```

```python
RGB_img = resized_RGB_img
red = plt.hist(RGB_img[:, :, 0].ravel(), bins = 256, color = 'Red', alpha = 0.5, density=True)
green = plt.hist(RGB_img[:, :, 1].ravel(), bins = 256, color = 'Green', alpha = 0.5, density=True)
blue = plt.hist(RGB_img[:, :, 2].ravel(), bins = 256, color = 'Blue', alpha = 0.5, density=True)
triple = plt.hist(RGB_img.ravel(), bins = 256, color = 'orange', alpha = 0.5, density=True) 
plt.legend(['Красный канал (R)', 'Зеленый канал (G)', 'Cиний канал (B)' , 'Сумма по каналама'])
plt.xlabel('Интенсивность $n$')
plt.ylabel('Количество')
plt.show()
print("для красного", RGB_img[:, :, 0].max)
print('Введите наибольшее значение плотности среди трех каналов:', float('{:.4f}'.format(max(red[0].max(), green[0].max(), blue[0].max()))))
```
