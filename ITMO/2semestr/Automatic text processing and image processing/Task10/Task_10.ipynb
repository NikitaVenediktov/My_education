```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
#Task 1 
img = cv2.imread('img_1.jpg')
RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(RGB_img)
```

```python
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(gray_img, cmap='gray')
```

```python
gray_hist = plt.hist(gray_img.ravel(),256,[0,256])
plt.title('Histogram for gray scale picture')
plt.xlabel('Интенсивность $n$')
plt.ylabel('Количество $h(n)$')
plt.show()
x_max = np.argmax(gray_hist[0]) 
y_max = gray_hist[0].max()
print('Введите максимальное значение функции', y_max, '\n', 'Введите интенсивность , при которой достигается максимум ', x_max)
```

```python
# Task 4 
img = cv2.imread('img_4.jpg') # меняем картинку
RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(RGB_img)
```

```python
RGB_img_new = cv2.GaussianBlur(RGB_img, (3,7), cv2.BORDER_DEFAULT)
plt.imshow(RGB_img_new)
```

```python
# Task 5 
img = cv2.imread('img_5.jpg') # меняем картинку
plt.imshow(img)
```

```python
#RES_ROOT = Path('/content/data/')
def contraharmonic_mean(img, size, Q):
    num = np.power(img, Q + 1)
    denom = np.power(img, Q)
    kernel = np.full(size, 1.0)
    result = cv2.filter2D(num, -1, kernel) / cv2.filter2D(denom, -1, kernel)
    return result


img = cv2.imread(('img_5.jpg'), cv2.IMREAD_GRAYSCALE)
plt.imshow(contraharmonic_mean(img, (3,3), 0.5), cmap='gray')
```
