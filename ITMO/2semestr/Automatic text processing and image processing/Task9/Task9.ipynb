```python
#Task 3 

# ath = 'img/' 
# file_name = 'markus-winkler-wvii1ayGs3A-unsplash.jpg'
img = cv2.imread('../Task8/task92.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

print(type(img))
print(img.shape) 
plt.imshow(img)
plt.show()

h, w, c = img.shape
new_width = 720

k = float(w) / h
dim = (new_width, int(new_width / k))



resized_RGB_img = cv2.resize(img, dim, interpolation = cv2.INTER_AREA)

print('Отношение:', resized_RGB_img.shape[1]/resized_RGB_img.shape[0]) 
print('ширина:', resized_RGB_img.shape[1]) 
print('Высота:', resized_RGB_img.shape[0]) 
plt.imshow(resized_RGB_img)
print(resized_RGB_img.shape) # новая размерность 
plt.show()
# cv2.imwrite('/content/img/1markus-winkler-wvii1ayGs3A-unsplash.jpg', cv2.cvtColor(resized_RGB_img, cv2.COLOR_RGB2BGR))
cv2.imwrite('task92_rgb.jpg', cv2.cvtColor(resized_RGB_img, cv2.COLOR_BGR2RGB))
print('bytes',os.stat('task92_rgb.jpg').st_size) # скачиваем изображение
```
